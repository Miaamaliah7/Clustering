```python
mall = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
mall.head()
```

```python
mall = mall.drop (['Annual Income (k$)'], axis = 1)
mall.head()
```

```python
mall_x = mall.iloc[:, 2:5]
mall_x.head()
```

```python
sns.scatterplot(x='Age', y='Spending Score (1-100)', data=mall, s=100, color='red', alpha = 0.5)
```

```python
x_array = np.array(mall_x)
print(x_array)
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
kmeans = KMeans(n_clusters = 3, random_state=123)

kmeans.fit(x_scaled)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
mall["kluster"] = kmeans.labels_
mall.head()
```

```python
fig, ax = plt.subplots()
sct = ax.scatter(x_scaled[:,1], x_scaled[:,0], s = 150,
c = mall.kluster, marker = 'o', alpha = 0.5)
centers = kmeans.cluster_centers_
ax.scatter(centers[:,1], centers[:,0], c='red', s=200, alpha=0.5);
plt.title('Hasil Klustering Mall Customer')
plt.xlabel('Scaled Age')
plt.ylabel('Scaled Spending Score (1-100')
plt.show()
```
